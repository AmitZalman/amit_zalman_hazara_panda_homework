```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.DataFrame({
    'Order ID': [101, 102, 103, 104, 105],
    'Customer': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Amount': [250, 450, 300, 150, 500],
    'Payment Method': ['Card', 'Cash', 'Card', 'Cash', 'Card']
})

df = df.set_index("Order ID")
display(df)

# Tasks
# 1 Sort the DataFrame by Amount in ascending order
df_sorted_amount_asc = df.sort_values(by="Amount", ascending=True)
display(df_sorted_amount_asc)

# 2 Sort the DataFrame by two columns: Payment Method and Amount, both in descending order
df_sorted_amount_Payment_des = df.sort_values(by= ["Payment Method" , "Amount"], ascending= [False , False])
display(df_sorted_amount_Payment_des)
```

```python

# Create DataFrame
df = pd.DataFrame({
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female'],
    'Club': ['Science', 'Drama', 'Science', 'Sports', 'Drama', 'Science', 'Sports', 'Drama'],
    'Scholarship': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes']
})

display(df)

# Tasks
# 1 Count how many students by Gender
category_counts = df['Gender'].value_counts().reset_index(name='Count')
display(category_counts)

# 2 Count how many students with a scholarship (Scholarship == 'Yes'), grouped by Club and Gender
scholarship_yes = df[df['Scholarship'] == 'Yes']
scholarship_counts = scholarship_yes.groupby(['Club', 'Gender']).size().reset_index(name='Count')
display(scholarship_counts)
```

```python
# Create DataFrame
df = pd.DataFrame([
    {'EmpID': 1, 'Name': 'Alice',   'Department': 'HR',        'Gender': 'Female', 'Sales': 1200, 'Bonus': 150},
    {'EmpID': 2, 'Name': 'Bob',     'Department': 'Finance',   'Gender': 'Male',   'Sales': 1800, 'Bonus': 200},
    {'EmpID': 3, 'Name': 'Charlie', 'Department': 'Finance',   'Gender': 'Male',   'Sales': 1700, 'Bonus': 180},
    {'EmpID': 4, 'Name': 'Diana',   'Department': 'HR',        'Gender': 'Female', 'Sales': 1600, 'Bonus': 175},
    {'EmpID': 5, 'Name': 'Evan',    'Department': 'IT',        'Gender': 'Male',   'Sales': 2200, 'Bonus': 250},
    {'EmpID': 6, 'Name': 'Fiona',   'Department': 'IT',        'Gender': 'Female', 'Sales': 2100, 'Bonus': 240},
    {'EmpID': 7, 'Name': 'George',  'Department': 'Finance',   'Gender': 'Male',   'Sales': 1950, 'Bonus': 210},
    {'EmpID': 8, 'Name': 'Hannah',  'Department': 'HR',        'Gender': 'Female', 'Sales': 1550, 'Bonus': 160},
    {'EmpID': 9, 'Name': 'Ian',     'Department': 'IT',        'Gender': 'Male',   'Sales': 2300, 'Bonus': 270},
    {'EmpID': 10,'Name': 'Julia',   'Department': 'Finance',   'Gender': 'Female', 'Sales': 1850, 'Bonus': 220},
])

df = df.set_index('EmpID')
display(df)

# Tasks
# 1 Group by Gender and print the count of employees
Gender_count_employees = df.groupby(['Gender']).size().reset_index(name='Count')
display(Gender_count_employees)

# 2 Group by Gender and print the mean of all numeric columns
gender_means = df.groupby('Gender').mean(numeric_only=True)
display(gender_means)

# 3 Group by Department and print the mean of all numeric columns
Department_means = df.groupby('Department').mean(numeric_only=True)
display(Department_means)

# 4 Group by ['Department', 'Gender'] and print the max, min, and std of the Bonus column
Department_Gender_means = df.groupby(['Department' , 'Gender'])['Bonus'].agg(['min','max','std'])
display(Department_Gender_means)


# 5 Print the list of unique values in the Department column
unique_values = df['Department'].unique()
print(unique_values)
```

```python
# Movie dataset with intentional duplicates
df = pd.DataFrame([
    {'movie_id': 1, 'title': 'Inception',        'rating': 9.0},
    {'movie_id': 2, 'title': 'Titanic',          'rating': 8.7},
    {'movie_id': 2, 'title': 'Titanic',          'rating': 8.7},  # full duplicate row
    {'movie_id': 3, 'title': 'Interstellar',     'rating': 8.9},
    {'movie_id': 4, 'title': 'Avatar',           'rating': 8.1},
    {'movie_id': 4, 'title': 'Avatar',           'rating': 8.1},  # full duplicate row
    {'movie_id': 4, 'title': 'Avatar',           'rating': 8.3},  # duplicate id only
    {'movie_id': 5, 'title': 'The Dark Knight',  'rating': 9.2}
])

display(df)
# Tasks
# 1 Print the fully duplicated rows (same movie_id, title, and rating)
duplicates = df[df.duplicated(subset=['movie_id', 'title', 'rating'], keep=False)]
print(duplicates)

# 2 Print the rows that are duplicates by movie_id only (even if rating differs)
duplicates_2 = df[df.duplicated(subset=['movie_id'], keep=False)]
print(duplicates_2)
```

```python

# 1 Keep only the first occurrence of each movie_id and drop all other duplicates.
print(df.drop_duplicates(subset=["movie_id"], keep="first"))
```

```python
# Product inventory dataset with NaN values
df = pd.DataFrame({
    "product": ["Laptop", np.nan, "Phone", "Tablet", "Camera"],
    "brand":   ["Dell", "Apple", "Samsung", np.nan, "Canon"],
    "price":   [1200, 1500, np.nan, 300, 700],
    "stock":   [10, np.nan, 25, 15, 8],
    "rating":  [4.5, 4.8, np.nan, 4.2, 4.6]
})

# Tasks
# 1 Display which cells are missing with df.isna().
display(df.isna())
# 2 Add a new column "number_of_missing" that counts how many NaN values each row has.
df['number_of_missing'] = df.isna().sum(axis=1) 
display(df)

# 3 Count how many missing values there are in the "price" column.
display(df['price'].isna().sum(axis=0))
```

```python
# Employees dataset with some missing values
df = pd.DataFrame({
    "first_name": ["Alice", np.nan, "Bob", "Diana", "Evan"],
    "last_name":  ["Smith", np.nan, "Brown", "Lopez", "Kim"],
    "age":        [28, np.nan, 35, 40, 30],
    "department": ["HR", "IT", "Finance", "IT", "HR"],
    "pre_training_score":  [7, np.nan, np.nan, 6, 8],
    "post_training_score": [9, np.nan, np.nan, 7, 9]
})
display(df)

# Find employees where pre_training_score is missing but the first_name is not missing
mising_training = df['pre_training_score'].isna() & df['first_name'].notna()
display(df[mising_training])
```

```python
# Products dataset with some NaN values
df = pd.DataFrame({
    "product": ["Laptop", "Phone", "Tablet", "Camera", "Monitor"],
    "brand":   ["Dell", np.nan, "Apple", "Canon", np.nan],
    "price":   [1200, 800, np.nan, 500, 300],
    "stock":   [10, 25, 15, np.nan, np.nan],
    "rating":  [4.5, np.nan, 4.8, 4.3, np.nan]
})

display(df)

# Add helper column: number of missing per row
df["number_of_missing"] = df.isna().sum(axis=1)
display(df)

# 1 Drop every row that has at least one missing value.
df.dropna()

# 2 Drop rows only if they have all values missing.
df = df.dropna(how = 'all')
display(df)

# 3 Drop rows that have fewer than 2 non-null values (thresh=2).
two_values = df.dropna(thresh = 2)
display(two_values)
```
